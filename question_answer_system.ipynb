### Preprocessing

```python
db_question=df['question']
```

```python
from nltk.stem import PorterStemmer,LancasterStemmer
ps = PorterStemmer()
lac= LancasterStemmer()
#stem_word=[]

from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords   #Import english stopword from nltk module
stop_words = set(stopwords.words('english'))
symbols = "!\"#$%&()*+--/:.';,<=>”?“@[\]^_`{|}~\n0123456789é£—‘"
def clean_sent(sent): #Input sentence
    
    val=[]
    words=sent.split(' ') #split sentence into words
    
     
    #filter_words=[]
    for word in words:
        if word in stop_words:      #If word is not in stopword than append it in filter_word
           # print('stopword is: ',word)
            words.remove(word)
        
        
    
    for each_word in words:        #for each word
        for alpha in each_word: #for each alphabet in word
            if alpha in symbols:
                each_word = each_word.replace(alpha,'')   #Remove symbols
        each_word = each_word.lower()
        each_word = each_word.strip()
        
        #Stemming
        each_word=ps.stem(each_word)
      #  each_word=lac.stem(each_word)
        
        #lemmatizing
        each_word = lem.lemmatize(each_word)
        
        if each_word!=chr(32) and len(each_word)>1:
            val.append(each_word)
    sentence = ' '.join(val)
    return sentence
```

```python
clean_question=[]
for sent in db_question:
    clean_question.append(clean_sent(sent))
    
#clean_question
```

### TF-IDF Model

```python
#tf_word=[]
def tf(sent):
    words=sent.split(' ')
    temp={}
    for word in words:
        tf=words.count(word)/len(words)
        temp[word]=tf
    #tf_word.append(temp)
    return temp
```

```python
tf_word=[]
for sent in clean_question:
    tf_word.append(tf(sent))
```

```python
tf_word
```

```python
count={}
def count_word(sent):
    words=sent.split(' ')
    for word in set(words):
        if word in count.keys():
            count[word]+=1
        else:
            count[word]=1
```

```python
for sent in clean_question:
    count_word(sent)
```

```python
import math
idf_value={}
def idf(allword,n):
    
    for word in allword:
        idf1=1+math.log(n/count[word])
        idf_value[word]=idf1
```
